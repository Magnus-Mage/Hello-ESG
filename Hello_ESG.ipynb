```python

%pip install transformers

from transformers import BertTokenizer, BertForSequenceClassification
import numpy as np

# Download and load the fine-tuned FinBERT sentiment classification model and the vocabulary (FinVocab, the finance vocabulary discussed earlier) from Hugging Face.
finbert_sent = BertForSequenceClassification.from_pretrained('yiyanghkust/finbert-tone', num_labels=3)  # fine-tuned FinBERT sentiment classification model
tokenizer_sent = BertTokenizer.from_pretrained('yiyanghkust/finbert-tone')  # the vocabulary (FinVocab, the finance vocabulary)

#################################################################

from transformers import BertTokenizer, BertForSequenceClassification, pipeline

finbert_esg = BertForSequenceClassification.from_pretrained('yiyanghkust/finbert-esg-9-categories',num_labels=9)
tokenizer_esg = BertTokenizer.from_pretrained('yiyanghkust/finbert-esg-9-categories')  # pre-train
nlp_esg = pipeline("text-classification", model=finbert_esg, tokenizer=tokenizer_esg)
```

```python

# %pip install pdfplumber
import pdfplumber
import pandas as pd
import re
from nltk import word_tokenize, sent_tokenize
import numpy as np

# Initialize your DataFrame
ESG_reports_df = pd.DataFrame(columns=[
    'company_id', 'E_neutral', 'E_positive', 'E_negative',
    'S_neutral', 'S_positive', 'S_negative',
    'G_neutral', 'G_positive', 'G_negative',
    'non_neutral', 'non_positive', 'non_negative'
])


for i, d in enumerate(pdf_file_names):
    text = ''


    part1, part2 = d.split('_')
    company_id = part1


    path = "ESG_report_110/" + d
    with pdfplumber.open(path) as pdf:
        for page in pdf.pages:
            text += page.extract_text()

    data_to_add = {'company_id': company_id}

    # Create a DataFrame from the new data
    new_data_df = pd.DataFrame([data_to_add])

    # Use pd.concat to add the new data
    ESG_reports_df = pd.concat([ESG_reports_df, new_data_df], ignore_index=True)

    # Initialize counts for sentiment and ESG categories
    ESG_reports_df.loc[i, 'E_neutral'] = 0
    ESG_reports_df.loc[i, 'E_positive'] = 0
    ESG_reports_df.loc[i, 'E_negative'] = 0
    ESG_reports_df.loc[i, 'S_neutral'] = 0
    ESG_reports_df.loc[i, 'S_positive'] = 0
    ESG_reports_df.loc[i, 'S_negative'] = 0
    ESG_reports_df.loc[i, 'G_neutral'] = 0
    ESG_reports_df.loc[i, 'G_positive'] = 0
    ESG_reports_df.loc[i, 'G_negative'] = 0
    ESG_reports_df.loc[i, 'non_neutral'] = 0
    ESG_reports_df.loc[i, 'non_positive'] = 0
    ESG_reports_df.loc[i, 'non_negative'] = 0


    def remove_special_characters(text):
        pattern = re.compile('[^a-zA-Z,.\!?]+')
        cleaned_text = pattern.sub(' ', text)
        return cleaned_text


    cleaned_text = remove_special_characters(text)

    if len(cleaned_text) == 0:
        continue

    for sent in sent_tokenize(cleaned_text):
        sent_r = sent[:400] if len(sent) > 400 else sent


        inputs = tokenizer_sent([sent_r], return_tensors="pt", padding=True, truncation=True, max_length=512)
        outputs = finbert_sent(**inputs)[0]

        # Assign each sentence to its most probable sentiment category.
        labels = {0: 'neutral', 1: 'positive', 2: 'negative'}

        sentiment = labels[np.argmax(outputs.detach().numpy()[0])]


        results_esg = nlp_esg([sent_r])
        esg_class_detail = results_esg[0]['label']

        if esg_class_detail in ['Climate Change', 'Natural Capital', 'Pollution & Waste']:
            esg_class = 'E'
        elif esg_class_detail in ['Human Capital', 'Product Liability', 'Community Relations']:
            esg_class = 'S'
        elif esg_class_detail in ['Corporate Governance', 'Business Ethics & Values']:
            esg_class = 'G'
        else:
            esg_class = 'non_ESG'


        if esg_class == 'E':
            ESG_reports_df.loc[i, f'E_{sentiment}'] += 1
        elif esg_class == 'S':
            ESG_reports_df.loc[i, f'S_{sentiment}'] += 1
        elif esg_class == 'G':
            ESG_reports_df.loc[i, f'G_{sentiment}'] += 1
        elif esg_class == 'non_ESG':
            ESG_reports_df.loc[i, f'non_{sentiment}'] += 1

ESG_reports_df.to_excel('ESG_sent_and_class_110.xlsx', index=True)
```

```python


# Define a function to calculate the ESG score for a company
def calculate_esg_score(row):
    # Define weights for sentiments (customize as needed)
    weight_positive = 1
    weight_neutral = 0.5
    weight_negative = -1

    # Calculate E, S, G scores
    E_score = (row['E_positive'] * weight_positive +
               row['E_neutral'] * weight_neutral +
               row['E_negative'] * weight_negative)

    S_score = (row['S_positive'] * weight_positive +
               row['S_neutral'] * weight_neutral +
               row['S_negative'] * weight_negative)

    G_score = (row['G_positive'] * weight_positive +
               row['G_neutral'] * weight_neutral +
               row['G_negative'] * weight_negative)

    return pd.Series({'E_score': E_score, 'S_score': S_score, 'G_score': G_score})

# Apply the function to each row
ESG_reports_df[['E_score', 'S_score', 'G_score']] = ESG_reports_df.apply(calculate_esg_score, axis=1)

# Calculate the total ESG score if needed
ESG_reports_df['Total_ESG_score'] = ESG_reports_df['E_score'] + ESG_reports_df['S_score'] + ESG_reports_df['G_score']
ESG_reports_df['Tone'] = (ESG_reports_df['E_positive'] + ESG_reports_df['S_positive'] + ESG_reports_df['G_positive'])/(ESG_reports_df['E_positive'] + ESG_reports_df['S_positive'] + ESG_reports_df['G_positive']+ESG_reports_df['E_neutral'] + ESG_reports_df['S_neutral'] + ESG_reports_df['G_neutral']+ESG_reports_df['E_negative'] + ESG_reports_df['S_negative'] + ESG_reports_df['G_negative'])

print(ESG_reports_df)
```

```python
#import what we need
from requests_html import AsyncHTMLSession
import nest_asyncio
import asyncio

# Apply nest_asyncio
nest_asyncio.apply()

#initialize AsyncHTMLSession
session = AsyncHTMLSession()

async def get_news():
    #use session to get the page
    r = await session.get('https://news.google.com/topstories?hl=en-GB&gl=GB&ceid=GB:en')

    #render the html, sleep=1 to give it a second to finish before moving on. scrolldown= how many times to page down on the browser, to get more results. 5 was a good number here
    await r.html.arender(sleep=1, scrolldown=5)

    #find all the articles by using inspect element and create blank list
    articles = r.html.find('article')
    newslist = []

    #loop through each article to find the title and link. try and except as repeated articles from other sources have different h tags.
    for item in articles:
        try:
            newsitem = item.find('h3', first=True)
            title = newsitem.text
            link = list(newsitem.absolute_links)[0]
            newsarticle = {
                'title': title,
                'link': link
            }
            newslist.append(newsarticle)
        except Exception as e:
            print(f"Error: {e}")
            pass

    #print the length of the list
    print(len(newslist))
    return newslist

#run the async function
news = asyncio.run(get_news())
```

```python
import json
from fpdf import FPDF

# Function to flatten nested JSON
def flatten_json(y):
    out = {}

    def flatten(x, name=''):
        if isinstance(x, dict):
            for a in x:
                flatten(x[a], name + a + '_')
        elif isinstance(x, list):
            i = 0
            for a in x:
                flatten(a, name + str(i) + '_')
                i += 1
        else:
            out[name[:-1]] = x

    flatten(y)
    return out

# Function to convert JSON to PDF
def json_to_pdf(json_file, pdf_file):
    # Read JSON data
    with open(json_file, 'r') as file:
        data = json.load(file)

    # Flatten JSON if it's nested
    if isinstance(data, list):
        data = [flatten_json(item) for item in data]
    else:
        data = [flatten_json(data)]

    # Extract headers
    headers = sorted({key for item in data for key in item.keys()})

    # Create a PDF object
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size=12)

    # Add a title to the PDF
    pdf.cell(200, 10, txt="JSON to PDF Conversion", ln=True, align='C')

    # Add a table header
    col_width = pdf.get_string_width(max(headers, key=len)) + 10
    for header in headers:
        pdf.cell(col_width, 10, header, 1, 0, 'C')
    pdf.ln()

    # Add table rows
    for item in data:
        for header in headers:
            text = str(item.get(header, ""))
            pdf.multi_cell(col_width, 10, text, 1, 'L', False)
            pdf.ln()

    # Save the PDF
    pdf.output(pdf_file)

# Specify the input JSON file and output PDF file
json_file = '/content/drive/MyDrive/Colab Notebooks/ESG_FinBERT/data14.json'
pdf_file = '/content/drive/MyDrive/Colab Notebooks/ESG_FinBERT/ESG_report_110/output14.pdf'  # Ensure this is a valid file path

# Convert JSON to PDF
json_to_pdf(json_file, pdf_file)
```

```python
def get_blog_url(soup):
    div_ = soup.find_all('div', attrs={'class': 'FL PR20'})
    url_list = []
    for title in div_:
        href = title.find('a')['href']
        url_list.append('https://www.moneycontrol.com/' + href)
    return url_list
```

```python
def get_blog_content(url):
    request = requests.get(url)
    soup = BeautifulSoup(request.text, 'html.parser')
    all_scripts = soup.find_all('script', attrs={'type': 'application/ld+json'})
    raw_article_str = all_scripts[2].get_text().replace('\n', ' ')
    parts = re.split(r'(["{,"}"])[^\n]*?}', raw_article_str)
    parts[1:-1] = map(lambda s: ''.join(s.split()), parts[1:-1])
    article_str = ''.join(parts)
    article_str = article_str[1:]
    article_str = article_str[:-1]
    article_dict = json.loads(article_str)
    all_tags = soup.find_all('div', attrs={'class': 'tags_first_line'})
    lst_all_tags = []
    for i in all_tags:
        lst_all_tags.append(i.get_text())
    tags = lst_all_tags[0].replace('TAGS:', '')
    tags = tags.replace('\n', ' ')
    tags = tags.split('#')
    tags = tags[1:]
    tags = ', '.join([str(elem).strip() for elem in tags])
    article_dict['tags'] = tags
    return article_dict
```

```python
def get_page_no(url, sc_id, page_no, next, year):
  request = request.get(url)
  soup = BeautifulSoup(request.text, 'html.parser')

  all_page_no = soup.find_all('div', attr = {'class': 'pages MR10 MT15'})
  page_list = [i.text for i in all_page_no[0].find_all('a')]

  if any(map(str.isdigit, page_list[-1])):
    return int(page_list[-1]), next

  else:
    next = next + 1
    page_no = int(page_list[-2])

    url = "https://www.moneycontrol.com/stocks/company_info/stock_news.php?sc_id="+sc_id+"&scat=&pageno="+str(page_no)+"&next="+str(next)+"&durationtype=Y&Year"+str(year)+"&duration=1&news_type="
    return get_page_no(url, sc_id, page_no, next, year)
```

```python
def save_company_data(url_="https://www.moneycontrol.com/stocks/company_info/stock_news.php?", sc_id=[], page_no=1, next=0, years=[]):
    for company in sc_id:
        df = pd.DataFrame(columns=['company', 'datePublished', 'author', 'headline',
                                   'description', 'articleBody', 'tags', 'url'])

        for year in years:
            print('year:', year)
            print('page_no:', page_no)
            print('next:', next)

            url = url_ + "sc_id=" + company + "&scat=&pageno=" + str(page_no) + "&next=" + str(next) + "&durationType=Y&year=" + str(year) + "&duration=1&news_type="
            print('url:', url)

            max_page_no, max_next = get_page_no(url, company, page_no, next, year)
            max_next = max_next + 1

            for i in range(max_next):
                for j in range((i*10)+1, (i*10)+11):
                    if j <= max_page_no:
                        url_list = []
                        url = url_ + "sc_id=" + company + "&scat=&pageno=" + str(j) + "&next=" + str(year) + "&duration=1&news_type="
                        request = requests.get(url)
                        soup = BeautifulSoup(request.text, 'html.parser')
                        url_list = get_blog_url(soup)

                        for url in url_list:
                            try:
                                article_dict = get_blog_content(url)
                                print(company)
                                print(article_dict['datePublished'])
                                print(article_dict['author'])
                                print(article_dict['headline'])
                                print(article_dict['description'])
                                print(article_dict['articleBody'])
                                print(article_dict['tags'])
                                print(article_dict['url'])
                                print('---------------------------------')

                                article_lst = [[company,
                                                article_dict['datePublished'],
                                                article_dict['author'],
                                                article_dict['headline'],
                                                article_dict['description'],
                                                article_dict['articleBody'],
                                                article_dict['tags'],
                                                article_dict['url']]]

                                df = df.append(pd.DataFrame(article_lst, columns=['company', 'datePublished', 'author', 'headline',
                                                                                  'description', 'articleBody', 'tags', 'url']), ignore_index=True)
                            except:
                                article_lst = [[company, 'error', 'error', 'error', 'error', 'error', 'error', url]]
                                df = df.append(pd.DataFrame(article_lst, columns=['company', 'datePublished', 'author', 'headline',
                                                                                  'description', 'articleBody', 'tags', 'url']), ignore_index=True)
                                continue
                        else:
                            break

            df.to_csv('/content/drive/MyDrive/Colab Notebooks/ESG_FinBERT/Json_Format/' + company + '.csv')
```
